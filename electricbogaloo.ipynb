```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import io 
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['after_wargame.csv']))

df.info()
```

```python
del df ["Description"]
```

```python
del df ["Unnamed: 0.1"]
```

```python
del df ["Unnamed: 0"]
```

```python
df['Interest'] = np.where(df['Players'] > df['Players'].mean(), "Alot", "Few")
df
```

```python
feature = ["Players"]
x = df[feature]
y = df["Interest"]
```

```python
svm = SVC(kernel='rbf', random_state = 1, gamma= 2, C=1)

svm.fit(x,y)

y_pred = svm.predict(x)
print(accuracy_score(y, y_pred))
```
